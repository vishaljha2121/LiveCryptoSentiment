```python
## Module import
import pandas as pd
import numpy as np 

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Data import
data_loc = '../Data/'
df = pd.read_csv(data_loc + 'dogecoin.csv')
```

```python
df.head()
```

```python
## Drop columns 
drop_cols = ['Unnamed: 0', 'id', 'conversation_id', 'created_at', 'place', 'cashtags', 'user_id', 'user_id_str', 'link', 'urls', 'photos', 'video', 'thumbnail', 'quote_url', 'search', 'near', 'geo', 'source', 'user_rt_id', 'user_rt', 'retweet_id', 'reply_to', 'retweet_date', 'translate', 'trans_src', 'trans_dest']
df.drop(drop_cols, axis=1, inplace=True)
```

```python
df.head()
```

```python
## DATA INFO
def data_info(data):
    display(data.head())
    print()

    display(data.tail())
    print()

    display(data.shape)
    print()

    display(data.info())
    print()

    try:
        display(data.select_dtypes(include=['object']).describe())
        print()
    except:
        print('No object data')
        print()
    finally:
        display(data.describe())
        print()
    
    display(data.isna().sum())
    print()

    total_missing_values = data.isna().sum().sum()
    total_cells = np.product(data.shape)

    print(f'Total missing values: {total_missing_values}')
    print(f'Total cells: {total_cells}')
    print(f'Percentage of missing values: {total_missing_values / total_cells * 100}')
```

```python
data_info(df)
```

```python
## Drop rows where language is not english
df = df[df['language'] == 'en']
df.head()
```

```python
## Drop cols
df.drop(['language', 'username', 'name', 'day', 'hour', 'retweet', 'hashtags'], axis=1, inplace=True)
df.head()
```

```python
## Function to seperate date and time from date column
def seperate_date_time(date):
    date_split = date.split(' ')
    date_ = date_split[0]
    time = date_split[1]
    return date_, time

df['date_'], df['time'] = zip(*df['date'].apply(seperate_date_time))
df.head()
```

```python
## Export data
df.to_csv(data_loc + 'dogecoin_cleaned.csv', index=False)
```
